```python
import pandas as pd

# Datos iniciales
data = {
    'Prioridad': [3, 3, 2, 1, 3, 2, 1, 3, 3, 2, 2, 3, 2, 1, 1],
    'Rubro': ['Informatica', 'Tecnología', 'Educacion', 'Corporaciones', 'Financiero', 'Retail', 'Ferias libres', 'Mineria', 'Industrias', 'Industrias transnacionales', 'Salud tecnico', 'Salud profesional', 'Aeronautica', 'Comunicación', 'Cooperativa']
}
df_prioridad = pd.DataFrame(data)

# Normalizar el indicador de prioridad
min_prioridad = df_prioridad['Prioridad'].min()
max_prioridad = df_prioridad['Prioridad'].max()

df_prioridad['Prioridad_Normalizada'] = (df_prioridad['Prioridad'] - min_prioridad) / (max_prioridad - min_prioridad)

# Mostrar el DataFrame con el indicador normalizado
df_prioridad = df_prioridad.sort_values(by='Prioridad_Normalizada', ascending=False)
df_prioridad
```

```python
# Definimos las categorías de renta y sus correspondientes categorías
income_categories = {
    550: ["Retail"],
    650: ["Retail", "Industrias", "Cooperativa"],
    800: ["Educacion", "Corporaciones", "Industrias", "Industrias transnacionales", "Salud tecnico", "Aeronautica", "Comunicación", "Cooperativa"],
    1100: ["Educacion", "Corporaciones", "Industrias", "Industrias transnacionales", "Salud tecnico", "Aeronautica", "Comunicación"],
    1400: ["Informatica", "Tecnología", "Educacion", "Corporaciones", "Financiero", "Mineria", "Salud profesional"],
    1600: ["Informatica", "Mineria", "Financiero", "Ferias libres", "Salud profesional"]
}

# Paso 1: Invertir el diccionario
professions = {}

for income, categories in income_categories.items():
    for category in categories:
        if category not in professions:
            professions[category] = []
        professions[category].append(income)

# Paso 2: Calcular el promedio de sueldos por profesión
average_salaries = {profession: sum(salaries) / len(salaries) for profession, salaries in professions.items()}

# Paso 3: Normalizar los valores
max_salary = max(average_salaries.values())
min_salary = min(average_salaries.values())

normalized_salaries = {profession: (avg_salary - min_salary) / (max_salary - min_salary) 
                       for profession, avg_salary in average_salaries.items()}

# Imprimir los resultados
for profession, normalized_salary in normalized_salaries.items():
    print(f"{profession}: {normalized_salary:.2f}")
```

```python
import pandas as pd

# Datos iniciales de prioridad y rubros
data_prioridad = {
    'Prioridad': [3, 3, 2, 1, 3, 2, 1, 3, 3, 2, 2, 3, 2, 1, 1],
    'Rubro': ['Informatica', 'Tecnología', 'Educacion', 'Corporaciones', 'Financiero', 'Retail', 'Ferias libres', 'Mineria', 'Industrias', 'Industrias transnacionales', 'Salud tecnico', 'Salud profesional', 'Aeronautica', 'Comunicación', 'Cooperativa']
}
df_prioridad = pd.DataFrame(data_prioridad)

# Normalizar el indicador de prioridad
min_prioridad = df_prioridad['Prioridad'].min()
max_prioridad = df_prioridad['Prioridad'].max()

df_prioridad['Prioridad_Normalizada'] = (df_prioridad['Prioridad'] - min_prioridad) / (max_prioridad - min_prioridad)

# Definimos las categorías de renta y sus correspondientes categorías
income_categories = {
    550: ["Retail"],
    650: ["Retail", "Industrias", "Cooperativa"],
    800: ["Educacion", "Corporaciones", "Industrias", "Industrias transnacionales", "Salud tecnico", "Aeronautica", "Comunicación", "Cooperativa"],
    1100: ["Educacion", "Corporaciones", "Industrias", "Industrias transnacionales", "Salud tecnico", "Aeronautica", "Comunicación"],
    1400: ["Informatica", "Tecnología", "Educacion", "Corporaciones", "Financiero", "Mineria", "Salud profesional"],
    1600: ["Informatica", "Mineria", "Financiero", "Ferias libres", "Salud profesional"]
}

# Paso 1: Invertir el diccionario
professions = {}

for income, categories in income_categories.items():
    for category in categories:
        if category not in professions:
            professions[category] = []
        professions[category].append(income)

# Paso 2: Calcular el promedio de sueldos por profesión
average_salaries = {profession: sum(salaries) / len(salaries) for profession, salaries in professions.items()}

# Paso 3: Normalizar los valores
max_salary = max(average_salaries.values())
min_salary = min(average_salaries.values())

normalized_salaries = {profession: (avg_salary - min_salary) / (max_salary - min_salary) 
                       for profession, avg_salary in average_salaries.items()}

# Convertir el diccionario a DataFrame para facilitar la unión
df_salaries = pd.DataFrame(list(normalized_salaries.items()), columns=['Rubro', 'Sueldo_Normalizado'])

# Unir la información de prioridades y sueldos normalizados
df_prioridad = df_prioridad.merge(df_salaries, on='Rubro')

# Datos iniciales de socios
data_socios = {
    'Cantidad socios': [3, 3, 3, 2, 2, 1],
    'Rango socios': ['70-120', '120-160', '160-200', '200-300', '300-400', '500 o más']
}
df_socios = pd.DataFrame(data_socios)

# Función para convertir el rango a punto medio
def rango_a_punto_medio(rango):
    if 'o más' in rango:
        return 500  # Tomar 500 como el punto medio para "500 o más"
    limites = rango.split('-')
    return (int(limites[0]) + int(limites[1])) / 2

# Aplicar la función a la columna 'Rango socios' para obtener el punto medio
df_socios['Punto medio'] = df_socios['Rango socios'].apply(rango_a_punto_medio)

# Normalizar los puntos medios
min_punto_medio = df_socios['Punto medio'].min()
max_punto_medio = df_socios['Punto medio'].max()

df_socios['Socios_Normalizado'] = (df_socios['Punto medio'] - min_punto_medio) / (max_punto_medio - min_punto_medio)

# Crear un diccionario para convertir rangos a valores normalizados
socios_normalizados_dict = {rango: normalizado for rango, normalizado in zip(df_socios['Rango socios'], df_socios['Socios_Normalizado'])}

# Función para calcular el puntaje del sindicato
def calcular_puntaje(fila, df_prioridad, socios_normalizados_dict):
    rubro = fila['Rubro']
    socios = fila['Cantidad socios']
    
    # Obtener la prioridad normalizada
    prioridad_normalizada = df_prioridad.loc[df_prioridad['Rubro'] == rubro, 'Prioridad_Normalizada'].values[0]
    
    # Obtener la normalización del número de socios
    if isinstance(socios, str):
        socios_normalizado = socios_normalizados_dict[socios]
    else:
        # Normalizar el número de socios si es un valor único
        socios_normalizado = (socios - min_punto_medio) / (max_punto_medio - min_punto_medio)
    
    # Calcular el puntaje como la media de las normalizaciones
    puntaje = (prioridad_normalizada + socios_normalizado) / 2
    
    return puntaje

# Ejemplo de uso
fila_ejemplo_1 = {'Rubro': 'Salud tecnico', 'Cantidad socios': 100}
fila_ejemplo_2 = {'Rubro': 'Salud tecnico', 'Cantidad socios': 500}
puntaje_1 = calcular_puntaje(fila_ejemplo_1, df_prioridad, socios_normalizados_dict)
puntaje_2 = calcular_puntaje(fila_ejemplo_2, df_prioridad, socios_normalizados_dict)
print(f"Puntaje del sindicato con 100 socios: {puntaje_1:.2f}")
print(f"Puntaje del sindicato con 500 socios: {puntaje_2:.2f}")
```

```python
import pandas as pd
import numpy as np

# Función para convertir el rango a punto medio
def rango_a_punto_medio(rango):
    if 'o más' in rango:
        return 500  # Tomar 500 como el punto medio para "500 o más"
    limites = rango.split('-')
    return (int(limites[0]) + int(limites[1])) / 2

# Función para calcular el puntaje del sindicato
def calcular_puntaje(fila, df_prioridad, socios_normalizados_dict):
    rubro = fila['Rubro']
    socios = fila['Cantidad socios']
    
    # Obtener la prioridad normalizada
    prioridad_filtrada = df_prioridad.loc[df_prioridad['Rubro'] == rubro, 'Prioridad_Normalizada']
    if not prioridad_filtrada.empty:
        prioridad_normalizada = prioridad_filtrada.values[0]
    else:
        prioridad_normalizada = 0  # O un valor por defecto como np.nan

    # Obtener la normalización del número de socios
    if socios in socios_normalizados_dict:
        socios_normalizado = socios_normalizados_dict[socios]
    else:
        # Normalizar el número de socios si es un valor único
        socios_normalizado = (socios - min_punto_medio) / (max_punto_medio - min_punto_medio)
    
    # Calcular el puntaje como la media de las normalizaciones
    puntaje = (prioridad_normalizada + socios_normalizado) / 2
    
    return puntaje

# Cargar los datos de sindicatos y categorías
df_sindicatos = pd.read_csv('data/todas_las_regiones.csv')
df_categorias = pd.read_excel('./data/CATEGORIAS.xlsx', sheet_name="SINDICATOS")

# Seleccionar la primera columna y la columna 5
df_categorias = df_categorias.iloc[:, [0, 5]]

# Eliminar la primera fila y la segunda fila se llama Rubro
df_categorias = df_categorias.drop([0])
df_categorias.rename(columns={'Unnamed: 5': 'Rubro'}, inplace=True)

# Cruzar df_sindicatos con df_categorias por la columna "nombre"
df_merge = pd.merge(df_sindicatos, df_categorias, on='nombre', how='left')

# Traer las filas que no tienen nan en la columna Rubro
df_merge = df_merge.dropna(subset=['Rubro'])

# Calcular la cantidad de socios
df_merge["Cantidad socios"] = df_merge["Socios"] + df_merge["Socias"]

# Normalización de la prioridad
data_prioridad = {
    'Prioridad': [3, 3, 2, 1, 3, 2, 1, 3, 3, 2, 2, 3, 2, 1, 1],
    'Rubro': ['Informatica', 'Tecnología', 'Educacion', 'Corporaciones', 'Financiero', 'Retail', 'Ferias libres', 'Mineria', 'Industrias', 'Industrias transnacionales', 'Salud tecnico', 'Salud profesional', 'Aeronautica', 'Comunicación', 'Cooperativa']
}
df_prioridad = pd.DataFrame(data_prioridad)
min_prioridad = df_prioridad['Prioridad'].min()
max_prioridad = df_prioridad['Prioridad'].max()
df_prioridad['Prioridad_Normalizada'] = (df_prioridad['Prioridad'] - min_prioridad) / (max_prioridad - min_prioridad)

# Normalización de socios
data_socios = {
    'Cantidad socios': [3, 3, 3, 2, 2, 1],
    'Rango socios': ['70-120', '120-160', '160-200', '200-300', '300-400', '500 o más']
}
df_socios = pd.DataFrame(data_socios)
df_socios['Punto medio'] = df_socios['Rango socios'].apply(rango_a_punto_medio)
min_punto_medio = df_socios['Punto medio'].min()
max_punto_medio = df_socios['Punto medio'].max()
df_socios['Socios_Normalizado'] = (df_socios['Punto medio'] - min_punto_medio) / (max_punto_medio - min_punto_medio)
socios_normalizados_dict = {rango: normalizado for rango, normalizado in zip(df_socios['Rango socios'], df_socios['Socios_Normalizado'])}

# Calcular el puntaje para cada sindicato
df_merge["Puntaje"] = df_merge.apply(calcular_puntaje, args=(df_prioridad, socios_normalizados_dict), axis=1)

# Ordenar por puntaje de forma descendente
df_merge = df_merge.sort_values(by='Puntaje', ascending=False)

# Mostrar el DataFrame resultante
df_merge.to_excel('sindicatos_puntaje.xlsx', index=False)
```
